```python
import requests
import json
user_key="user_key=INSERT_KEY_HERE"
fst_url = "https://api.crunchbase.com/v3.1/organizations?"+user_key
fst_querystring = {"locations":"Switzerland","organization_types":"company"}
companies = []
```

```python
def collect(js,arr):
    for comp in js['data']['items']:
        arr.append([comp['properties']['name'].strip(),comp['properties']['api_url'].strip(),comp['uuid'].strip(),comp['properties']['api_url'].strip(),comp.get('city_name'),comp.get('region_name')])

response = requests.request("GET", fst_url, params=fst_querystring)
js = json.loads(response.text)
collect(js,companies)

while js['data']['paging']['current_page'] < js['data']['paging']['number_of_pages']:
    next_url = json.dumps(js['data']['paging']['next_page_url'],indent=2)[1:-1]+"&"+user_key
    next_response = requests.request("GET", next_url)
    js = json.loads(next_response.text)
    print("Page "+str(js['data']['paging']['current_page'])+"...")
    collect(js,companies)

print("Done!")
```

```python
import json
cmps = json.loads(open('companies-with-location.json').read())
```

```python
# filter companies without fouding date.

cmps2 = [c for c in cmps if c['founded_on']!=None]
```

```python
award_years = [2018,2017,2016,2015,2014,2013]
eligibles_by_year = dict()

from datetime import datetime 
def parse_date(s):
    return datetime.strptime(s,'%Y-%m-%d')

def is_eligible_by_founding_date(date,award_year): # eligible ranges from august of award year to janauary of 5 years before
    after_end = parse_date(str(award_year)+"-09-01")
    before_start = parse_date(str(award_year-6)+"-12-31")
    return before_start < date and date < after_end

for y in award_years:
    mylist = [c for c in cmps2 if is_eligible_by_founding_date(parse_date(c['founded_on']),y)]
    eligibles_by_year[str(y)] = mylist
```
